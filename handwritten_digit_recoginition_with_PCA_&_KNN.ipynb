```python
import pandas as pd 
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
```

```python
data_set_test=pd.read_csv("./dataset/mnist_test.csv")
data_set_train=pd.read_csv("./dataset/mnist_train.csv")
```

```python
dataset=pd.concat([data_set_test,data_set_train],axis=0,ignore_index=True)
dataset.sample(10)
```

```python
print("Dataset train : ",data_set_train.shape)
print("Dataset test : ",data_set_test.shape)
print("Combined Dataset : ",dataset.shape)
```

```python
X=dataset.iloc[:,1:]
```

```python
X.sample(5)
```

```python
# apply scalling on the dataset
scaler=StandardScaler()
X_Scaled=scaler.fit_transform(X)
```

```python
# comparison between scaled and unscaled X
print("----------Un-Scaled------------")
print(X)
print("------------Scaled------------")
print(X_Scaled)
```

```python
# spilt the dataset 
Y=dataset['label']
x_train,x_test,y_train,y_test=train_test_split(X_Scaled,Y,random_state=True,train_size=0.8)
```

```python
print("X-Train",x_train.shape)
print("Y-Train",y_train.shape)
print("X-Test",x_test.shape)
print("Y-Test",y_test.shape)
```

```python
# Train PCA to give N principle Components
pca=PCA(n_components=100)
n_components=pca.fit_transform(x_train)

n_components
```

```python
knn=KNeighborsClassifier(n_neighbors=5)

knn.fit(n_components,y_train)
```

```python

components=pca.transform(x_test)


prediction=knn.predict(components)
```
